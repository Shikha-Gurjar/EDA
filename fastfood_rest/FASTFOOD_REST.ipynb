## 4.6  Restaurents in  Top 3 cities 

```python
fast_city = fastnew.set_index(['city','name'])
fast_city_alpha = fast_city.sort_index(inplace = False)
fast_city_alpha.xs('Cincinnati', level = 'city')    # through this can find its province and any other type of details
city_cincinnati = fast_city_alpha.groupby(['city']).get_group('Cincinnati')
value_city_cin  = city_cincinnati.groupby(['name']).count()
value_city_cin
value_city_cin.plot.line(figsize = (20,5),yticks = np.arange(0,20,5))
plt.title('Restaurents in Cincinnati' , fontsize =20 )
```

__Observation__

__McDonald's__ has __19__ Restaurents in __Cincinnati__. following __Wendy's__.

```python
fast_city_alpha.xs('Las Vegas', level = 'city')
city_lasv = fast_city_alpha.groupby(['city']).get_group('Las Vegas')
value_city_lasv  = city_lasv.groupby(['name']).count()
value_city_lasv.plot.bar(figsize = (20,5),yticks = np.arange(0,10,1),rot = 90)
plt.title('Restaurents in Las Vegas' , fontsize =20 )
```

__Observation__

__Mcdonald's__ have maximum no of restaurents. Following __Del Taco, Domino's pizza, Pizza Hut.__ 

```python
fast_city_alpha.xs('Houston', level = 'city')
city_houston = fast_city_alpha.groupby(['city']).get_group('Houston')
value_city_houston  = city_houston.groupby(['name']).count()
value_city_houston
value_city_houston.plot.bar(figsize = (20,5),yticks = np.arange(0,12,1), rot =90)
plt.title('Restaurents in Houston' , fontsize =20 )
```

__Observation__

From above graph we got that the highest number of restaurents  are of "__McDonald's__". After McDonald's __Burger King__ and __Whataburger__ are there

## 4.9  Location using latitude and longitudes

As we know that Latitude is defined with respect to an equatorial reference plane. IF the latitude is positiveit is  northerly if  the latitude is negative it will be southerly . 

And Longitude is defined in terms of meridians, east of the prime meridian, the longitude is positive; and the west of the prime meridian, the longitude is negative.

### 4.9.1 Through Equatorial Plane

#### 4.9.1.1 In  North Equatorial Plane

```python
fastnew[fastnew['latitude']>0].plot.hexbin(x = 'latitude' , y = 'longitude', gridsize = 9)
plt.title("NORTH EQUATORIAL PLANE ",fontsize = 16)
```

__Observation__

The __hexabin__ graph showing the locations which are the __North side of equatorial plane__. __9998__ entries  are here from this dataset.
